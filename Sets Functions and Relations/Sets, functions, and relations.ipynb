Plot each of the following functions from $\mathbb{Z}$ to $\mathbb{Z}$ over a representative subdomain (use points and not a line since the domain is $\mathbb{Z}$).

a) $f(n) = n-1$

b) $f(n) = n^3$

c) $f(n) = n^2 + 1$

```python
import matplotlib.pyplot as plt
a = [] #Initialize an empty array for domain
b = [] #Initialize an empty array for codomain
for n in range(-10, 10):
  fn = n-1      # Given function
  a.append(n)   #Append all domain values
  b.append(fn)  #Append all codomain values
print(a)
print(b)
fig , ax = plt.subplots(1,1, figsize= (5,5))
plt.scatter(a , b)#Plot all points in domain and codomain
plt.show()
```

a) $f(n) = n-1$ For $n$ to be all integers from $-\infty \ to \infty$, the range of $$f(n)=(n-1)\ is -\infty \ to \infty$$
The function is injective as for every element in domain we have only 1 output element in co-domain as (n-1) will give unique outputs for positive and negative integers.
Also, it is surjective as for any element in codomain (n-1) there is only one unique n in domain.
Hence, function is bijective as well.

### 6. Composition

a) Suppose that $f(x) = x^2 + 1$ and $g(x) = e^x$ are functions from $\mathbb{R}$ into $\mathbb{R}$. Find $f \circ g$ and $g \circ f$ and plot them in Python (on the same axis to compare).

b) Suppose $f(x) = a x + b$ and $g(x) = c x + d$ are functions from $\mathbb{R}$ into $\mathbb{R}$. Find constants $a$, $b$, $c$ and $d$ such that $f \circ g = g \circ f$. 

```python
import matplotlib.pyplot as plt
import math
import numpy as np
def composite(f, g):
  """Performs a composition of two functions f(x) and g(x)"""
  return lambda x : f(g(x))
def F(x):
  """Performs the given function: x^2 + 1"""
  return x**2 + 1
def G(x):
  """Performs the given function: exp(x)"""
  return math.exp(x)

fog = composite(F , G) #Performs F(G(x))
gof = composite(G , F) #Performs G(F(x))

#Plotting the two functions in a single plot
fig,ax = plt.subplots(1, 1, figsize = (10,10))
ax.set_title('FoG and GoF')
for x in np.linspace(-1.5,1.5,1000):
  ax.plot(x , fog(x), "o", label= 'FoG', color= 'blue')
  ax.plot(x , gof(x), "o", label= 'GoF', color= 'red')
```

### 7. Relations

Define the relation $R$ on $\mathbb{N} \times \mathbb{N}$ saying that $x \; R\; y$ if and only if the binary representations of $x$ and $y$ have the same number of ones. For example, $15 \; R \; 23$ since $15$ in binary is $1111$ and $23$ in binary is $10111$, which both have four ones. 

Show that $R$ is an equivalence relation.
 

```python
def dec_bin(x): #To convert decimal to binary
    if x == 0: 
      return [0]
    b = []
    while x:
      b.append(x % 2)
      x = x//2
    return b[::-1]

def count(B): #Count number of 1s in the binary number
  N = 0
  for i in range(len(B)):
    if B[i] == 1:
      N = N+1
  return N

num1 = int(input('Enter number 1:'))
num2 = int(input('Enter number 2:'))
print(dec_bin(num1))
print(dec_bin(num2))
n= 1000
#To check if number of 1s in both binary numbers are equal
R = {(X,Y) for X in range(1 , n) for Y in range(1 , n) if count(dec_bin(X))==count(dec_bin(Y))}
print('XRY=', (num1,num2) in R)

#To check relation is reflexive
def ref(R):
  for a1,a2 in R:
    if a1==a2 and (a1,a2) in R:
      return True
  return False

#To check if relation is symmetric
def sym(R):
  for a1,a2 in R:
    for b1,b2 in R:
      if a1==b2 and a2==b1:
        return True
  return False

#To check if relation is transitive
def tran(R):
 for a1,a2 in R:
  for b1,b2 in R:
   if a2==b1 and (a1,b2) not in R:
      return False
  return True

Ref = ref(R)
Sym = sym(R)
Tran = tran(R)
#To check if relation is equivalence relation
if(Ref, Sym, Tran == True):
  print("Relation is equivalence")
```
